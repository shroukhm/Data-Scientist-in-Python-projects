RealAgents sells a variety of types of house in one metropolitan area.

Some houses sell slowly and sometimes require lowering the price in order to find a buyer.

In order to stay competitive, RealAgents would like to optimize the listing prices of the houses it is trying to sell.

They want to do this by predicting the sale price of a house given its characteristics.

If they can predict the sale price in advance, they can decrease the time to sale.

# Task 2 

Before you fit any models, you will need to make sure the data is clean. 

The table below shows what the data should look like. 

Create a cleaned version of the dataframe. 

 - You should start with the data in the file "house_sales.csv". 

 - Your output should be a dataframe named `clean_data`. 

 - All column names and values should match the table below.


| Column Name | Criteria                                                |
|-------------|---------------------------------------------------------|
| house_id    | Nominal. </br> Unique identifier for houses. </br>Missing values not possible. |
| city        | Nominal. </br>The city in which the house is located. One of 'Silvertown', 'Riverford', 'Teasdale' and 'Poppleton' </br>Replace missing values with "Unknown". |
| sale_price  | Discrete. </br>The sale price of the house in whole dollars. Values can be any positive number greater than or equal to zero.</br>Remove missing entries. |
| sale_date   | Discrete. </br>The date of the last sale of the house. </br>Replace missing values with 2023-01-01. |
| months_listed  | Continuous. </br>The number of months the house was listed on the market prior to its last sale, rounded to one decimal place. </br>Replace missing values with mean number of months listed, to one decimal place. |
| bedrooms    | Discrete. </br>The number of bedrooms in the house. Any positive values greater than or equal to zero. </br>Replace missing values with the mean number of bedrooms, rounded to the nearest integer. |
| house_type   | Ordinal. </br>One of "Terraced", "Semi-detached", or "Detached". </br>Replace missing values with the most common house type. |
| area      | Continuous. </br>The area of the house in square meters, rounded to one decimal place. </br>Replace missing values with the mean, to one decimal place. |

```python
# Import necessary libraries
import pandas as pd

# Load the dataset from a CSV file into a DataFrame
house_data = pd.read_csv('house_sales.csv')

# Task 2: Clean the dataset according to the provided criteria

# 1. Clean the 'city' column:
# Replace missing values (denoted as '--') with "Unknown"
house_data['city'] = house_data['city'].replace('--', 'Unknown')

# 2. Clean the 'sale_price' column:
# Remove rows where 'sale_price' is missing (NaN)
house_data = house_data[house_data['sale_price'].notna()]

# 3. Clean the 'sale_date' column:
# Replace missing values with the default date "2023-01-01"
house_data['sale_date'] = house_data['sale_date'].fillna('2023-01-01')

# 4. Clean the 'months_listed' column:
# Calculate the mean of 'months_listed' and replace missing values with this mean,
# rounded to one decimal place
mean_months_listed = house_data['months_listed'].mean()
house_data['months_listed'] = house_data['months_listed'].fillna(round(mean_months_listed, 1))

# 5. Clean the 'bedrooms' column:
# Calculate the mean of 'bedrooms' and replace missing values with this mean,
# rounded to the nearest integer
mean_bedrooms = house_data['bedrooms'].mean()
house_data['bedrooms'] = house_data['bedrooms'].fillna(round(mean_bedrooms))

# 6. Clean the 'house_type' column:
# Define a mapping for house type abbreviations to their full names
house_type_mapping = {
    'Terraced': 'Terraced',
    'Terr.': 'Terraced',
    'Semi-detached': 'Semi-detached',
    'Semi': 'Semi-detached',
    'Detached': 'Detached',
    'Det.': 'Detached'
}
# Map the 'house_type' column using the defined mapping
house_data['house_type'] = house_data['house_type'].map(house_type_mapping)

# Identify the most common house type and replace missing values with it
most_common_house_type = house_data['house_type'].mode()[0]
house_data['house_type'] = house_data['house_type'].fillna(most_common_house_type)

# 7. Clean the 'area' column:
# Remove the 'sq.m.' suffix, convert the values to float, and replace missing values with the mean,
# rounded to one decimal place
house_data['area'] = house_data['area'].str.replace('sq.m.', '').astype(float)
mean_area = house_data['area'].mean()
house_data['area'] = house_data['area'].fillna(round(mean_area, 1))

# Convert 'city' and 'house_type' columns to category type as they are nominal data
house_data['city'] = house_data['city'].astype('category')
house_data['house_type'] = house_data['house_type'].astype('category')

# Ensure the 'sale_date' column is in datetime format for better date handling
house_data['sale_date'] = pd.to_datetime(house_data['sale_date'], format='%Y-%m-%d')

# Output the cleaned DataFrame as 'clean_data' for further analysis
clean_data = house_data

# Display the first few rows of the cleaned DataFrame to verify the changes
print(clean_data.head())
```

# Task 3 

The team at RealAgents have told you that they have always believed that the number of bedrooms is the biggest driver of house price. 

Producing a table showing the difference in the average sale price by number of bedrooms along with the variance to investigate this question for the team.

 - You should start with the data in the file 'house_sales.csv'.

 - Your output should be a data frame named `price_by_rooms`. 

 - It should include the three columns `bedrooms`, `avg_price`, `var_price`. 

 - Your answers should be rounded to 1 decimal place.   

```python
import pandas as pd  # Import the pandas library for data manipulation and analysis

# Load the dataset from a CSV file into a DataFrame named 'house_data'
house_data = pd.read_csv('house_sales.csv')

# Group the data by the number of bedrooms and calculate the average and variance of the sale prices
price_by_rooms = house_data.groupby('bedrooms')['sale_price'].agg(
    avg_price='mean',  # Calculate the average sale price for each bedroom count
    var_price='var'    # Calculate the variance of the sale price for each bedroom count
).reset_index()  # Reset the index to get a clean DataFrame structure

# Round the calculated average sale price to 1 decimal place
price_by_rooms['avg_price'] = price_by_rooms['avg_price'].round(1)

# Round the calculated variance of the sale price to 1 decimal place
price_by_rooms['var_price'] = price_by_rooms['var_price'].round(1)

# Display the resulting DataFrame containing average and variance of sale prices by bedroom count
print(price_by_rooms)
```

# Task 4

Fit a baseline model to predict the sale price of a house.

 1. Fit your model using the data contained in “train.csv” </br></br>

 2. Use “validation.csv” to predict new values based on your model. You must return a dataframe named `base_result`, that includes `house_id` and `price`. The price column must be your predicted values.

```python
import pandas as pd  # Import pandas for data manipulation
from sklearn.linear_model import LinearRegression  # Import LinearRegression model from sklearn

# Step 1: Load training data
train_data = pd.read_csv('train.csv')  # Read the training data from a CSV file into a DataFrame

# Step 2: Prepare the data for modeling
# Selecting all features except the target variable 'sale_price'
X_train = train_data.drop(columns=['sale_price'])  # Drop 'sale_price' column to get feature variables
y_train = train_data['sale_price']  # Store the target variable 'sale_price' separately

# Convert categorical variables to dummy variables (one-hot encoding)
X_train = pd.get_dummies(X_train, drop_first=True)  # Create dummy variables for categorical features and drop the first to avoid multicollinearity

# Fit the baseline model
model = LinearRegression()  # Initialize the linear regression model
model.fit(X_train, y_train)  # Fit the model using the training data

# Step 3: Load validation data
validation_data = pd.read_csv('validation.csv')  # Read the validation data from a CSV file into a DataFrame

# Prepare the validation data
X_val = validation_data  # Start with the validation data

# Convert categorical variables to dummy variables (one-hot encoding) for validation data
X_val = pd.get_dummies(X_val, drop_first=True)  # Create dummy variables for categorical features in validation data

# Align validation data with training data (ensuring both DataFrames have the same columns)
X_val = X_val.reindex(columns=X_train.columns, fill_value=0)  # Reindex to match the training features; fill missing columns with 0

# Step 4: Predicting new values
predictions = model.predict(X_val)  # Use the fitted model to predict sale prices based on validation features

# Step 5: Create the base_result DataFrame
base_result = pd.DataFrame({
    'house_id': validation_data['house_id'],  # Keep 'house_id' for the output to identify each prediction
    'price': predictions  # Store predicted prices
})

# Display the base_result DataFrame
print(base_result)  # Output the DataFrame containing house IDs and their predicted prices
```

# Task 5

Fit a comparison model to predict the sale price of a house.

 1. Fit your model using the data contained in “train.csv” </br></br>

 2. Use “validation.csv” to predict new values based on your model. You must return a dataframe named `compare_result`, that includes `house_id` and `price`. The price column must be your predicted values.

```python
import pandas as pd  # Import pandas for data manipulation
from sklearn.ensemble import RandomForestRegressor  # Import RandomForestRegressor from sklearn

# Step 1: Load training data
train_data = pd.read_csv('train.csv')  # Read the training data from a CSV file into a DataFrame

# Prepare the training data
X_train = train_data.drop(columns=['sale_price'])  # Drop the target variable 'sale_price' to get features
y_train = train_data['sale_price']  # Store the target variable 'sale_price' separately

# Convert categorical variables to dummy variables (one-hot encoding)
X_train = pd.get_dummies(X_train, drop_first=True)  # Create dummy variables for categorical features, dropping the first category to avoid multicollinearity

# Step 2: Fit the comparison model
comparison_model = RandomForestRegressor(random_state=104)  # Initialize the RandomForestRegressor with a fixed random seed for reproducibility
comparison_model.fit(X_train, y_train)  # Fit the model using the training data

# Step 3: Load validation data
validation_data = pd.read_csv('validation.csv')  # Read the validation data from a CSV file into a DataFrame

# Prepare the validation data
X_val = validation_data  # Start with the validation data

# Convert categorical variables to dummy variables (one-hot encoding) for validation data
X_val = pd.get_dummies(X_val, drop_first=True)  # Create dummy variables for categorical features in validation data

# Ensure the validation set has the same features as the training set
X_val = X_val.reindex(columns=X_train.columns, fill_value=0)  # Reindex to match training features; fill missing columns with 0

# Step 4: Predicting new values
predictions = comparison_model.predict(X_val)  # Use the trained model to predict sale prices based on validation features

# Step 5: Create the compare_result DataFrame
compare_result = pd.DataFrame({
    'house_id': validation_data['house_id'],  # Include 'house_id' from validation data to identify each prediction
    'price': predictions  # Store predicted prices
})

# Display the compare_result DataFrame
print(compare_result)  # Output the DataFrame containing house IDs and their predicted prices
```
